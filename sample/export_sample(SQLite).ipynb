# 最終課題・csvデータ取得サンプルコード

最終課題のデータを取得するための参考資料です。
以下を実行することで、2つのテーブルより、全てのデータを取得することができます。
**※ 以下、SQLiteがローカルにインストールされている場合のターミナルでの実行方法となります。**

sqlite3.exeが配置されているディレクトリに、最終課題で配布されたデータベースファイルファイル「telecom.sqlite3」を置いてください。
このディレクトリで、ターミナル（コマンドプロンプト）を起動し、次のようにデータベースに接続します。

```python
sqlite3 telecom.sqlite3
```

以下、プロンプト sqlite> が表示されていることを確認してから、実行してください。

```python
# 各種設定内容（デフォルト）の表示 - 適宜実行すると変更内容が確認できます
.show
```

```python
# テーブル名の確認
.tables
```

```python
# 出力の表示形式を変更
.mode csv
```

```python
# 区切り文字の変更
.separator ","
```

```python
# ヘッダーとしてカラム名を表示
.headers ON
```

```python
# 出力先に任意のcsvファイルを指定
.output [任意のファイル名1].csv
```

```python
# Clientテーブルから全データを取得
select * from Client;
```

```python
# 出力先に任意のcsvファイルを指定
.output [任意のファイル名2].csv
```

```python
# Recordテーブルから全データを取得
select * from Record;
```

```python
.exit
```
